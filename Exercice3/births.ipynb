```python
import pandas as pd

births = pd.read_csv('births.csv')

births.describe(include="all")
```

## Nettoyage des donnees

la fonction `dropna` permet de supprimer les lignes qui contiennent des `NaN`.

```python
births_drop = births.dropna()

births_drop.describe(include="all")
```

```python
births_drop.pivot_table('births', index='year', columns='gender', aggfunc="sum")
```

```python
decade = pd.cut(births_drop['year'], [1960, 1970, 1980, 1990], right=False)

print(decade)

births_drop.pivot_table('births', index=['year', decade], columns='gender', aggfunc="sum")
```
