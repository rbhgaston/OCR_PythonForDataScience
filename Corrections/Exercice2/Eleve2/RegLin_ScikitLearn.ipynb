# Effectuer une regresson linéaire avec Scikit-Learn

## Import des librairies nécessaires

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

## Lecture des données

```python
hubble = pd.read_csv('hubble_data.csv')
hubble.head() #On visualise le début de la table
```

## Visualisation des données

```python
fig = plt.figure(figsize = (10,7))
ax = plt.axes()
plt.plot(hubble['distance'], hubble['recession_velocity'], 'ro', markersize = 4)
ax = ax.set(xlabel='distance', ylabel='recession_velocity')
plt.title("Loi d'Hubble", fontsize = 25);
```

## Régression linéaire

```python
#On met les données au bon format
X, y = np.array(hubble['distance']).reshape(-1,1), np.array(hubble['recession_velocity']).reshape(-1,1)

#Fit avec la régression linéaire
regr = linear_model.LinearRegression()
regr.fit(X, y)

#Graphique représentant les données et la droite approximée
plt.figure(figsize=(10,7))
plt.xlabel('distance')
plt.ylabel('recession_velocity')
beta = float(regr.coef_) #Coefficient directeur 
alpha = float(regr.intercept_) #Ordonnée à l'origine
t = np.linspace(0,2.5,20)
plt.scatter(X, y, color = 'blue', marker = 'o')
plt.plot(t, beta*t+alpha, linestyle='--', color='black')
plt.title("Loi d'Hubble", fontsize = 25)
plt.show();

#Calcul du coefficient de corrélation avec la méthode score
R = regr.score(X, y)
print("Le coefficient de corrélation R² est :", R)
print("Le coefficient directeur et l'ordonnée à l'origine sont respectivements égaux à", beta, "et", alpha)
```
