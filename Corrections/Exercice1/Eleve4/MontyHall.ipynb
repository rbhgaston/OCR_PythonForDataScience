```python
%matplotlib inline
from matplotlib import pyplot as plt
from monty import generate_monty_hall
```

```python
samples = 10000
with_changement = generate_monty_hall(True, samples)
without_changement = generate_monty_hall(False, samples)
```

```python
plot = plt.bar(
    [1, 2],
    [sum(with_changement), sum(without_changement)], 
    tick_label=["Change","Keep"]
)
```

```python
change_wins = []
keep_wins = []
samples =  [1000, 10000, 20000, 50000, 80000, 100000]
for s in samples:
    change_wins.append(sum(generate_monty_hall(True, s)))
    keep_wins.append(sum(generate_monty_hall(False, s)))
```

```python
figure = plt.figure()
plot = plt.scatter(samples, change_wins)
plot = plt.scatter(samples, keep_wins)
```
