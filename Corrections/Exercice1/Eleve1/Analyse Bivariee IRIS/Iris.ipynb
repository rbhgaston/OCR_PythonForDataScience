```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as sp
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Chargement
iris = pd.read_csv("iris_dataset.csv")
```

```python
# On renomme les colonnes
iris.columns = ["id","sepal_length","sepal_width","petal_length","petal_width","species"]
```

```python
# On supprime l'identifiant des iris
del iris["id"]
```

```python
# On supprime les individus contenant au moins une valeur manquante
iris_dna = iris.dropna(axis=0, how='any')
print("iris : {} individus, iris_dna : {} individus".format(len(iris),len(iris_dna)))
```

```python
# Affichage des diagrammes de dispersion
sns.pairplot(iris_dna,hue="species")
plt.show()
```

```python
iris_setosa = iris_dna[iris_dna["species"] == "setosa"]
iris_virginica = iris_dna[iris_dna["species"] == "virginica"]
iris_versicolor = iris_dna[iris_dna["species"] == "versicolor"]
```

# CAS 1:Relation entre la longueur et la largeur du pétale, sur l'ensemble

```python
Data1=np.array(iris_dna)
X = np.reshape(Data1[:,2],(146,1))
y=np.reshape(Data1[:,3],(146,1))
```

```python
reg1= LinearRegression().fit(X,y)
print(reg1)
print("score",reg1.score(X, y))

print("coef",reg1.coef_)

print("intercept",reg1.intercept_) 


plt.scatter(X,y)
Y=reg1.coef_*X+reg1.intercept_
plt.plot(X,Y,color='y')
plt.xlabel('Longueur du pétale')
plt.ylabel('Largeur du pétale')
plt.title('Iris:Relation entre la largeur et la longueur du pétale')
plt.grid(True)
```


# CAS 2 : Relation entre la largeur du pétale et celle du sépale/ Setosa

```python
ARRAY_iris_setosa = np.array(iris_setosa)
ARRAY_iris_setosa.shape
```

```python
X2 = np.reshape(ARRAY_iris_setosa[:,3],(49,1))
y2=np.reshape(ARRAY_iris_setosa[:,1],(49,1))
reg2 = LinearRegression().fit(X2,y2)
print(reg2)
print("score",reg2.score(X2, y2))

print("coef",reg2.coef_)

print("intercept",reg2.intercept_) 


plt.scatter(X2,y2)
Y2=reg2.coef_*X2+reg2.intercept_
plt.plot(X2,Y2,color='y')
plt.xlabel('Largeur du pétale')
plt.ylabel('Largeur du sépale')
plt.title('Iris Setosa:Relation entre la largeur du sépale et celle du pétale')
plt.grid(True)
```

# Question Q4

```python
coeffs = {
    "cas 1" : {'a': [reg1.coef_] , 'b':[reg1.intercept_]},
    "cas 2" : {'a': [reg2.coef_] , 'b':[reg2.intercept_]},
    "cas 3" : {'a': [reg3.coef_] , 'b':[reg3.intercept_]},
    "cas 4" : {'a': [reg4.coef_] , 'b':[reg4.intercept_]},
}
lignes_modifiees = []

for (i,individu) in iris.iterrows(): # pour chaque individu de iris,...
    if pd.isnull(individu["petal_width"]): #... on test si individu["petal_width"] est nul.
        a = coeffs["cas 1"]['a']
        b = coeffs["cas 1"]['b']
        X = individu["petal_length"]
        Y = np.multiply(a,X) + b
        iris.loc[i,"petal_width"] = Y # on remplace la valeur manquante par Y
        lignes_modifiees.append(i)
        print("On a complété petal_width par {} a partir de petal_length={}".format(Y,X))
        
    if pd.isnull(individu["sepal_width"]):
        espece = individu["species"]
        X = individu["petal_width"]
        if espece=='setosa':
            a = coeffs["cas 2"]['a']
            b = coeffs["cas 2"]['b']
            Y = np.multiply(a,X) + b
        elif espece=='virginica':
            a = coeffs["cas 3"]['a']
            b = coeffs["cas 3"]['b']
            Y = np.multiply(a,X) + b
        else:
            a = coeffs["cas 4"]['a']
            b = coeffs["cas 4"]['b']
            Y = np.multiply(a,X) + b
        iris.loc[i,"sepal_width"] = Y
        lignes_modifiees.append(i)
        print("On a complété sepal_width par {} a partir de l'espece:{} et de petal_width={}".format(Y,espece,X))
```
