```python
import pandas as pd
data = pd.read_csv("libr_3.txt.csv")
data
```

```python
import numpy as np
#nettoyage des mois de naissance invalides;les de plus de 12 sont 
#éliminés
VALID_MONTH = np.arange(1,13, 1)          
def check_month(month):

    if month not in VALID_MONTH:

        print(' - "{}" n\'est pas un mois valide, nous le supprimons.' \

            .format(month))

        return np.NaN

    return month
#nettoyage des années de naissance invalides 
VALID_YEAR = np.arange(1969,2009, 1)          
def check_year(year):

    if year not in VALID_YEAR:

        print(' - "{}" n\'est pas une année valide, nous le supprimons.' \

            .format(year))

        return np.NaN

    return year
#nettoyage des jours de naissances invalides
VALID_DAYS = np.arange(1,32, 1)          
def check_day(day):

    if day not in VALID_DAYS:

        print(' - "{}" n\'est pas un jour valide, nous le supprimons.' \

            .format(day))

        return np.NaN

    return day
data['month'] = data['month'].apply(check_month)
data['month']= data['month']
data['year'] = data['year'].apply(check_year)
data['year']= data['year']
data['day'] = data['day'].apply(check_day)
data['day']= data['day']
print(data)
```
