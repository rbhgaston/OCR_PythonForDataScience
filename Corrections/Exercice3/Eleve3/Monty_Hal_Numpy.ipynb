# Monty Hall avec NumPy

```python
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt

# On importe le module qui contient l'algorithme du Monty Hall
from monty_hall_game import play, Strategie
```

## Construction des tableaux

```python
gains_changer = []
gains_garder = []
gains_rand = []
samples =  [1000, 10000, 20000, 50000, 80000, 100000]
for tours in samples:
    gains_changer.append(play(Strategie.CHANGER, tours))
    gains_garder.append(play(Strategie.GARDER, tours))
    gains_rand.append(play(Strategie.RANDOM, tours))
```

## Visualisatioin des gains en fonction du nombre de tours pour chaque strat√©gie

```python
figure = plt.figure()
plot = plt.scatter(samples, [x.sum() for x in gains_changer])
plot = plt.scatter(samples, [x.sum() for x in gains_garder])
plot = plt.scatter(samples, [x.sum() for x in gains_rand])
```

## Visualisation des moyennes de gains

```python
plot_mean = plt.scatter(samples, [r.mean() for r in gains_changer])
plot_mean = plt.scatter(samples, [r.mean() for r in gains_garder])
plot_mean = plt.scatter(samples, [r.mean() for r in gains_rand])
```
