```python
import numpy as np
import matplotlib.pyplot as plt

from monty_hall import Strategie, play

nb_tours = 100000

gains_changer = play(Strategie.GARDER, nb_tours)
gains_garder = play(Strategie.CHANGER, nb_tours)

print("En changeant de porte, le joueur a gagné {} sur {} parties."
      .format(np.sum(gains_changer), nb_tours))

print("En gardant son choix initial, le joueur a gagné {} sur {} parties."
      .format(np.sum(gains_garder), nb_tours))
```

# Visualisation des résultats

```python
%matplotlib inline

nb_tours = 10

figure = plt.figure()
ax = plt.axes()

plot = plt.plot(play(Strategie.CHANGER, nb_tours))

# titre du graphique
plt.title("Resultats de la simulation ({} parties)".format(nb_tours))

# label des axes
ax = ax.set(xlabel='# tour', ylabel='resultat')
```

```python
nb_tours = 10000

figure = plt.figure()
ax = plt.axes()

plot = plt.plot(play(Strategie.CHANGER, nb_tours))

# titre du graphique
plt.title("Resultats de la simulation ({} parties)".format(nb_tours))

# label des axes
ax = ax.set(xlabel='# tour', ylabel='resultat')
```

```python
nb_tours = 10000

figure = plt.figure()
ax = plt.axes()

plot = plt.scatter(range(nb_tours), play(Strategie.CHANGER, nb_tours))

# titre du graphique
plt.title("Resultats de la simulation ({} parties)".format(nb_tours))

# label des axes
ax = ax.set(xlabel='# tour', ylabel='resultat')
```

```python
nb_tours = 10000

figure = plt.figure()
ax = plt.axes()

plot = plt.bar([1,2],[np.sum(play(Strategie.CHANGER, nb_tours)), 
               np.sum(play(Strategie.GARDER, nb_tours))], 
        tick_label=["Changer","Garder"])

# titre du graphique
plt.title("Histogramme ({} parties)".format(nb_tours))

# label des axes
ax = ax.set(xlabel='strategie', ylabel='gain')
```

```python
gains_changer = []
gains_garder = []

samples =  [1000, 10000, 20000, 50000, 80000, 100000]

for tours in samples:
    gains_changer.append(play(Strategie.CHANGER, tours))
    gains_garder.append(play(Strategie.GARDER, tours))

figure = plt.figure()
ax = plt.axes()

plot = plt.scatter(samples, [sum(x) for x in gains_changer])
plot = plt.scatter(samples, [sum(x) for x in gains_garder])

# titre du graphique
plt.title("Gains des joueurs en fonction du nombre de tours")

# label des axes
ax = ax.set(xlabel='nombre de tours', ylabel='gain')
```
